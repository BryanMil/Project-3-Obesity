```python
#Dataframe creation by Bryan Miller
#Clean and Merge National Obesity by State and covid deaths 
import pandas as pd
from pathlib import Path
import numpy as np
from pprint import pprint
import sqlite3

csv_path = Path("../Resources/covid_confirmed_usafacts.csv")
Confirmed_df = pd.read_csv(csv_path, encoding="utf-8")

csv_path1 = Path("../Resources/covid_deaths_usafacts.csv")
Deaths = pd.read_csv(csv_path1, encoding="utf-8")

csv_path1 = Path("../Resources/states.csv")
States = pd.read_csv(csv_path1, encoding="utf-8")
```

```python
Confirmed_df.head()
conf_df = Confirmed_df[['State','1/22/2020','6/1/2020','1/1/2021','6/1/2021','1/1/2022','6/1/2022','1/1/2023']]
conf_df
```

```python
conf_df1 = conf_df.groupby(conf_df['State']).sum()


# conf_df1.rename(columns={'State':'Abbreviation'}, inplace=True)
# conf_df1.head()
# # C_df = conf_df1.reset_index()
# # C_df.head()
```

```python
conf_df1.rename(columns={'State':'Abbreviation'}, inplace=True)
conf_df1.head()
```

```python
States.head()
```

```python
States.rename(columns={'State':'St','Abbreviation':'State'}, inplace=True)
States.head()
```

```python
Conf_merge = States.merge(conf_df1,on='State')
Conf_merge.head()
```

```python
Final_Conf = Conf_merge[['St','1/22/2020','6/1/2020','1/1/2021','6/1/2021','1/1/2022','6/1/2022','1/1/2023']]
Final_Conf.head()
```

```python
Final_Conf.to_csv("../Resources/Final_Confirmed1.csv",
                  encoding="utf-8", index=False, header=True)
```

```python
Deaths.head()
```

```python
Deaths1 = Deaths[['State','1/22/2020','6/1/2020','1/1/2021','6/1/2021','1/1/2022','6/1/2022','1/1/2023']]
Deaths1
```

```python
Deaths2 = Deaths1.groupby(Deaths1['State']).sum()
Deaths2.head()
```

```python
States.head()
```

```python
Deaths3 = States.merge(Deaths2,on='State')
Deaths3.head()
```

```python
Deaths4 = Deaths3[['St','1/22/2020','6/1/2020','1/1/2021','6/1/2021','1/1/2022','6/1/2022','1/1/2023']]
Deaths4.head()
```

```python
Deaths4.to_csv("../Resources/Final_Deaths.csv",
                  encoding="utf-8", index=False, header=True)
```
